```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"D:\PythonProjects\StudentStressAnalisys\Student Stress Factors.csv")
df.head()
```

```python
df.info()
```

### Sleep Quality vs. Stress Levels:



```python
sleep_quality = df['Kindly Rate your Sleep Quality üò¥']
stress_levels = df['How would you rate your stress levels?']

# Calculate the correlation between Sleep Quality and Stress Levels
correlation = sleep_quality.corr(stress_levels)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(sleep_quality, stress_levels, color='blue', alpha=0.6)
plt.title('Sleep Quality vs. Stress Levels')
plt.xlabel('Sleep Quality')
plt.ylabel('Stress Levels')
plt.grid(True)

# Display the correlation coefficient on the plot
plt.text(1, 5, f'Correlation: {correlation:.2f}', fontsize=12, color='red')

# Show the plot
plt.show()
```

### Headache Frequency:



```python
# Select the column for headache frequency
headache_frequency = df['How many times a week do you suffer headaches ü§ï?']

# Calculate the average headache frequency
average_frequency = headache_frequency.mean()

# Create a bar chart to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(headache_frequency, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Headache Frequency')
plt.xlabel('Headache Frequency (times per week)')
plt.ylabel('Count')

# Display the average frequency on the chart
plt.axvline(average_frequency, color='red', linestyle='dashed', linewidth=2, label=f'Average Frequency: {average_frequency:.2f}')
plt.legend()

# Show the plot
plt.show()
```

### Time-based Analysis:

### Frequency of Responses:



```python

# Convert the "Timestamp" column to a datetime object
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Count the number of responses for each unique timestamp
response_counts = df['Timestamp'].value_counts().sort_index()

# Create a line plot to visualize the response frequency over time
plt.figure(figsize=(12, 6))
response_counts.plot(marker='o', color='skyblue')
plt.title('Frequency of Responses Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Number of Responses')
plt.grid
```
